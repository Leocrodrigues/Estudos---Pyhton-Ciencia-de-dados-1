```python
# calculado a temperatura media da semana
#ordenando os dias
temp_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']
temp_por_dia_da_semana = temp_por_dia_da_semana[nome_dias]
temp_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temp_por_dia_da_semana))

eixo.bar(indice, temp_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dias da semana', fontsize=15)
eixo.set_ylabel('Temperatura Média', fontsize=15)
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temp_por_dia_da_semana))

eixo.bar(indice, temp_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dias da semana', fontsize=15)
eixo.set_ylabel('Temperatura Média', fontsize=15)

# Colocando o nome dos dias no eixo X

eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
# Grafico de pizza

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
# Grafico de pizza
# colocando porcentagem
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
# Grafico de pizza
# colocando porcentagem
# separando(explodindo) alguns dias
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0,0,0,0,0.1,0.1,0.1]
#obs 0.1 e a distância podendo ser alterado


eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
# Grafico de pizza
# colocando porcentagem
# separando(explodindo) alguns dias
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0,0,0,0,0.1,0.1,0.1]
#obs 0.1 e a distância podendo ser alterado


eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
fig = plt.figure(figsize=(15,8))
eixo= fig.add_axes([0,0,1,1])

# criando dicionário atribuindo a cada planta uma cor e um marcador
cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

# criando uma função para ler os dados e atribuir cada cor ao tipo de planta 
for especie in df.espécie.unique():
    tmp =df[df['espécie'] ==  especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color = cores[especie], marker=marcadores[especie])

eixo.set_title('Gráfico de dispersão', fontsize=15, pad=15)

eixo.set_xlabel('Comprimento da Sépala', fontsize=15)
eixo.set_ylabel('Largura da Sépala', fontsize=15)

eixo.legend(cores, fontsize=15)

# aumento o tamanho dos dados nos eixos
eixo.tick_params(labelsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo= fig.add_axes([0,0,1,1])

# criando dicionário atribuindo a cada planta uma cor e um marcador
cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

# criando uma função para ler os dados e atribuir cada cor ao tipo de planta 
# colocando os marcadores
# e aumentando o tamanho com s
for especie in df.espécie.unique():
    tmp =df[df['espécie'] ==  especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color = cores[especie], marker=marcadores[especie],
                s=100)

eixo.set_title('Gráfico de dispersão', fontsize=15, pad=15)

eixo.set_xlabel('Comprimento da Sépala', fontsize=15)
eixo.set_ylabel('Largura da Sépala', fontsize=15)

eixo.legend(cores, fontsize=15)

# aumento o tamanho dos dados nos eixos
eixo.tick_params(labelsize=15)
```
